## Environment Setup

```python
import snscrape.modules.twitter as sntwitter

import pandas as pd
```

## Scraping For 2020

```python
# Creating list to append tweet data to
tweets_list_scrape = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('resesi ekonomi 2023 since:2020-02-15 until:2020-12-31 lang:in').get_items()):
    tweets_list_scrape.append([tweet.date, tweet.id, tweet.content, tweet.user.username, tweet.lang])
```

```python
# Creating a dataframe from the tweets list above
tweets_df_scrape = pd.DataFrame(tweets_list_scrape, columns=['Datetime', 'Tweet Id', 'Tweet', 'Username', 'Language'])

# Display first 5 entries from dataframe
tweets_df_scrape.head(100)
```

```python
# Export dataframe into a CSV
tweets_df_scrape.to_csv('tweet_scrape_result_2020.csv', sep=',', index=False)
```
